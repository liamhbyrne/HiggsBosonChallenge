## Imports

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

# Data Exploration

```python
df = pd.read_csv('training.csv')
df
```

## Comparing signal and background data

```python
df_without_index = df.drop(["EventId", "Weight"], axis=1)
```

```python
backgrounds = df_without_index.loc[df['Label'] == 'b']
"Row count: {}".format(len(backgrounds.index))
```

```python
signals = df_without_index.loc[df['Label'] == 's']
"Row count: {}".format(len(signals.index))
```

```python
plt.figure()
sns.distplot(backgrounds['DER_mass_MMC'], label="Background")
sns.distplot(signals['DER_mass_MMC'], label="Signal")
plt.legend()
plt.title("The estimated mass mH of the Higgs boson candidate")
```

### Comparing how each column is distributed for signals vs background

```python
for col in df_without_index.drop('Label', axis=1).columns: # Plot a distplot of each column excluding the label column
    fig, ax = plt.subplots()
    sns.distplot(backgrounds[col], label="Background")
    sns.distplot(signals[col], label="Signal")
    plt.legend()
    plt.show()
```
